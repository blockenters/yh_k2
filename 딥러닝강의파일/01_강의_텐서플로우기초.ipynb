# The Hello World of Deep Learning with Neural Networks

y = (2 * x) - 1 


```
float hw_function(float x){
    float y = (2 * x) - 1;
    return y;
}
```




## Imports



````
import tensorflow as tf
import numpy as np
from tensorflow import keras
````

## Define and Compile the Neural Network

1개의 레이어, 그리고 1개의 뉴런을 가지고 있는 뉴럴 네트워크 모델링 하기

````
model = tf.keras.Sequential( [  tf.keras.layers.Dense(units= 1, input_shape = (1,) ) ] )
````

컴파일 하기 위해서는 2개의 함수가 필요하다. 

loss 와 optimizer.

학습을 통해 예측한 값과, 실제의 y=10x+10 값 두개를 측정하는 LOSS function 을 설정해야 한다.

위의 로스를 최소화 하기 위해서 OPTIMIZER function 을 설정한다.

로스펑션으로 'MEAN SQUARED ERROR'를 사용하고, 옵티마이저로 'STOCHASTIC GRADIENT DESCENT' 를 사용한다. 아직 수학적으로 알 필요는 없다.



````
model.compile(optimizer='sgd', loss='mse')
````

## Providing the Data

y=2x-1 이 식에 따라, x의 데이터 6개와, 그에 해당하는 y의 데이터 6개를 준비한다. 머신러닝은 데이터로부터 학습하는 것이기 때문이다. 이 데이터로부터 우리는 위의 식을 컴퓨터가 찾아내도록 하는것이다. 

'Numpy' 라이브러리를 이용해서, xs 와 ys 를 만든다. 각각 6개씩

````
x = np.array( [-1, 0, 1, 2, 3, 4] )
````

````
y = np.array([-3, -1, 1, 3, 5, 7])
````

# Training the Neural Network

x와 y의 관계를 찾기 위해서, 학습하는 함수인 **model.fit** 함수를 호출한다. 각 에포크별로 좋은 모델인지 아닌지 평가할 수 있는 loss 값을 제공한다. 

````
model.fit(x, y, epochs=500)
````

모델의 학습이 끝나면 이제 **model.predict** 함수를 이용해서 새로운 데이터를 예측할 수 있다. 10을 넣어보자. 무슨 값이 예측되었는가?

````
model.predict([56])
````

````
# y = 2x -1 
2 * 56 - 1
````

19가 정확히 나오지 않은 이유는, 뉴럴넷은 확률로 처리한다. 따라서 주어진 데이터 6개를 보고, 식이 Y=2X-1 과 비슷할 것이라고 예측한 것이다. 그래서 거의 19라고 예측을 해준다.
